# **Batsman DataSet**

***Virat Kohli is an Indian cricketer. He was born in Delhi, India on November 5, 1988. Virat is the first player in ICC cricket history to win all 3 ICC awards in a single year- ICC ODI player of the year, ICC Test player of the year and ICC Player of the year award in 2018
Analyze the batting pefomance of ViraT Kohli. [Dataset : Batsman.csv]***

***LOADING THE DATASET***

```python
import pandas as pd
df = pd.read_csv('/kaggle/input/virat-kohli-batsman-dataset/Batsman.csv')
print(df.head(10))
print("\n")
print(df.info())
```

# ***Batsman’s overall batting performance***

```python
total_runs = df['RunsScored'].sum()
total_balls = df['BallsFaced'].sum()
total_innings = df.shape[0]
average_runs = total_runs / total_innings
strike_rate = (total_runs / total_balls) * 100

print(f"Total Runs: {total_runs}")
print(f"Total Innings: {total_innings}")
print(f"Batting Average (runs per innings): {average_runs:.2f}")
print(f"Strike Rate: {strike_rate:.2f}")
```

# ***Total Runs Per Year***

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Runs per year
plt.figure(figsize=(12,6))
sns.barplot(x='Year', y='RunsScored', data=df, estimator=sum, ci=None, color='skyblue')
plt.title("Virat Kohli: Total Runs per Year")
plt.xlabel("Year")
plt.ylabel("Total Runs")
plt.show()
```

# ***Runs Scored by Batting Position***

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='BattingPosition', y='RunsScored', data=df)
plt.title("Runs Scored by Batting Position")
plt.xlabel("Batting Position")
plt.ylabel("Runs Scored")
plt.show()
```

# ***Virat Kohli: Dismissal Types***

```python
dismissal_counts = df['Dismissal'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=dismissal_counts.index, y=dismissal_counts.values)
plt.title("Virat Kohli: Dismissal Types")
plt.xlabel("Dismissal")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# ***Virat Kohli: Total Runs vs Opposition***

```python
# Runs against top opponents
runs_vs_opp = df.groupby('Opposition')['RunsScored'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=runs_vs_opp.index, y=runs_vs_opp.values, palette='coolwarm')
plt.title("Virat Kohli: Total Runs vs Opposition")
plt.xlabel("Opposition")
plt.ylabel("Runs Scored")
plt.show()
```

# ***Virat Kohli: Strike Rate by Inning***

```python
df['StrikeRate'] = (df['RunsScored'] / df['BallsFaced']) * 100
plt.figure(figsize=(12,6))
sns.boxplot(x='Inning', y='StrikeRate', data=df)
plt.title("Virat Kohli: Strike Rate by Inning")
plt.xlabel("Inning")
plt.ylabel("Strike Rate")
plt.show()
```
